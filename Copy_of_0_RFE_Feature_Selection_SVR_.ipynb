**Input and Output Preparation**

```python
X = df2.drop(['pIC50'],axis = 1)
X
```

```python
Y = df2.pIC50
Y
```

**Initiation of Recursive Feature Elimination (RFE) for Support Vector Regressor**

```python
svr = LinearSVR(max_iter=1000)
sel = RFE(estimator = svr, n_features_to_select=50)
sel.fit(X, Y)
```
